```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from IPython.display import display
from tqdm.notebook import tqdm


def build_simple_model(embedding_dim=768):
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=(None, embedding_dim)),
        Dropout(0.2),
        LSTM(64),
        Dropout(0.2),
        Dense(256, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer=Adam(0.001), loss='mse', metrics=['mae'])
    return model


def prepare_sequences(df_news, df_target, window_minutes=30):
    """
    Create sequences where each minute's volatility is predicted
    using news from the previous 30 minutes.
    """
    sequences = []
    targets = []
    timestamps = []

    target_minutes = df_target.resample('1min')['BTCUSDT_target'].first().dropna()

    print(f"Processing {len(target_minutes)} minutes...")

    for i, (current_time, target_vol) in enumerate(target_minutes.items()):
        window_start = current_time - pd.Timedelta(f'{window_minutes}min')

        news_in_window = df_news[
            (df_news.index >= window_start) &
            (df_news.index < current_time)
        ]

        if len(news_in_window) == 0:
            continue

        try:
            news_embeddings = np.vstack(news_in_window['embedding'].values)
            sequences.append(news_embeddings)
            targets.append(target_vol)
            timestamps.append(current_time)
        except:
            continue

    print(f"Created {len(sequences)} training samples")
    return sequences, np.array(targets), timestamps


def train_model(model, sequences, targets, test_split=0.2, epochs=1):
    """Train model and return predictions."""
    split_idx = int(len(sequences) * (1 - test_split))

    train_seq = sequences[:split_idx]
    train_targets = targets[:split_idx]
    test_seq = sequences[split_idx:]
    test_targets = targets[split_idx:]

    print(f"Training on {len(train_seq)} samples, testing on {len(test_seq)} samples")

    for epoch in range(epochs):
        losses = []
        train_preds_epoch = []
        train_actuals_epoch = []
        with tqdm(total=len(train_seq), desc=f"Epoch {epoch+1}/{epochs}") as pbar:
            for seq, target in zip(train_seq, train_targets):
                seq_input = seq.reshape(1, len(seq), seq.shape[1])
                target_input = np.array([[target]])

                loss = model.train_on_batch(seq_input, target_input)
                losses.append(loss[0] if isinstance(loss, list) else loss)

                train_pred = model.predict(seq_input, verbose=0)[0][0]
                train_preds_epoch.append(train_pred)
                train_actuals_epoch.append(target)


                pbar.update(1)

        avg_loss = np.mean(losses)
        r2_epoch = r2_score(train_actuals_epoch, train_preds_epoch)

        print(f"Epoch {epoch+1}/{epochs}, Loss: {avg_loss:.4f}, Train R²: {r2_epoch:.4f}")

    predictions = []
    with tqdm(total=len(test_seq), desc="Predicting") as pbar:
        for seq in test_seq:
            seq_input = seq.reshape(1, len(seq), seq.shape[1])
            pred = model.predict(seq_input, verbose=0)[0][0]
            predictions.append(pred)
            pbar.update(1)


    return np.array(predictions), test_targets


def evaluate_and_plot(predictions, actuals, timestamps):
    """Calculate metrics and create plots."""
    mse = mean_squared_error(actuals, predictions)
    rmse = np.sqrt(mse)
    r2 = r2_score(actuals, predictions)

    print(f"\nResults:")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²: {r2:.4f}")

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

    n_show = min(100, len(timestamps))
    ax1.plot(timestamps[-n_show:], actuals[-n_show:], label='Actual', alpha=0.8)
    ax1.plot(timestamps[-n_show:], predictions[-n_show:], label='Predicted', alpha=0.8)
    ax1.set_title(f'Last {n_show} Minutes')
    ax1.set_ylabel('Volatility')
    ax1.legend()
    ax1.tick_params(axis='x', rotation=45)

    ax2.scatter(actuals, predictions, alpha=0.6)
    min_val, max_val = min(actuals.min(), predictions.min()), max(actuals.max(), predictions.max())
    ax2.plot([min_val, max_val], [min_val, max_val], 'r--')
    ax2.set_xlabel('Actual')
    ax2.set_ylabel('Predicted')
    ax2.set_title('Predictions vs Actuals')

    plt.tight_layout()
    plt.show()

    return {'RMSE': rmse, 'R²': r2}


def predict_volatility(df_news, df_target, window_minutes=30, epochs=5):
    """
    Main function: Predict minute-by-minute volatility using 30-minute news windows.

    Args:
        df_news: DataFrame with news embeddings, datetime index
        df_target: DataFrame with volatility targets, datetime index
        window_minutes: Minutes of news history to use (default: 30)
        epochs: Training epochs

    Returns:
        dict: Evaluation metrics
    """
    print(f"Predicting volatility every minute using {window_minutes}-minute news windows")

    sequences, targets, timestamps = prepare_sequences(df_news, df_target, window_minutes)

    if not sequences:
        print("No valid sequences found!")
        return None

    embedding_dim = sequences[0].shape[1]
    model = build_simple_model(embedding_dim)

    predictions, test_targets = train_model(model, sequences, targets, epochs=epochs)
    test_timestamps = timestamps[int(len(timestamps) * 0.8):]

    metrics = evaluate_and_plot(predictions, test_targets, test_timestamps)

    print(f"\nPredicted volatility for {len(predictions)} minutes")
    return metrics


metrics = predict_volatility(df_news, df_target, window_minutes=30, epochs=3)
```

```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.preprocessing.sequence import pad_sequences

class AttentionAggregator(tf.keras.layers.Layer):
    """Learn how to aggregate variable-length sequences"""

    def __init__(self, hidden_dim=128, **kwargs):
        super().__init__(**kwargs)
        self.hidden_dim = hidden_dim

    def build(self, input_shape):
        self.W_attention = self.add_weight(
            shape=(input_shape[-1], self.hidden_dim),
            initializer='random_normal',
            trainable=True,
            name='attention_weights'
        )

        self.v_attention = self.add_weight(
            shape=(self.hidden_dim, 1),
            initializer='random_normal',
            trainable=True,
            name='attention_vector'
        )

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[2])

    def call(self, inputs, mask=None):
        attention_hidden = tf.nn.tanh(tf.matmul(inputs, self.W_attention))
        attention_scores = tf.matmul(attention_hidden, self.v_attention)
        attention_scores = tf.squeeze(attention_scores, axis=-1)

        if mask is not None:
            mask_penalty = (tf.cast(mask, attention_scores.dtype) - 1.0) * 1e4  # Reduced from 1e9
            attention_scores += mask_penalty

        attention_scores = tf.clip_by_value(attention_scores, -10.0, 10.0)

        attention_weights = tf.nn.softmax(attention_scores, axis=1)

        attention_weights = attention_weights + 1e-8
        attention_weights = attention_weights / tf.reduce_sum(attention_weights, axis=1, keepdims=True)

        weighted_sum = tf.reduce_sum(
            inputs * tf.expand_dims(attention_weights, -1),
            axis=1
        )

        return weighted_sum

def build_model(embedding_dim=768, num_target_variables=10, attention_hidden_dim=256):
    """Build attention-based model that can handle variable-length sequences"""
    model = tf.keras.Sequential([
        tf.keras.layers.Masking(mask_value=0.0),
        AttentionAggregator(hidden_dim=attention_hidden_dim),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(num_target_variables)
    ])

    optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001, clipnorm=1.0)
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])
    return model

def prepare_data_for_attention(df_news, df_target, window_minutes=30, max_sequence_length=None):
    """Prepare padded sequences for attention model using sliding window approach"""
    sequences = []
    targets = []
    timestamps = []

    target_minutes = df_target.resample('1min')['BTCUSDT_target'].first().dropna()

    print(f"Processing {len(target_minutes)} minutes...")

    for i, (current_time, target_vol) in enumerate(target_minutes.items()):
        window_start = current_time - pd.Timedelta(f'{window_minutes}min')

        news_in_window = df_news[
            (df_news.index >= window_start) &
            (df_news.index < current_time)
        ]

        if len(news_in_window) == 0:
            continue

        try:
            news_embeddings = np.vstack(news_in_window['embedding'].values)
            sequences.append(news_embeddings)
            targets.append(target_vol)
            timestamps.append(current_time)
        except:
            continue

    print(f"Created {len(sequences)} training samples")

    if max_sequence_length is None:
        max_sequence_length = max(len(seq) for seq in sequences)

    padded_sequences = []
    for seq in sequences:
        if len(seq) > max_sequence_length:
            padded_seq = seq[:max_sequence_length]
        else:
            pad_length = max_sequence_length - len(seq)
            padded_seq = np.vstack([seq, np.zeros((pad_length, seq.shape[1]))])
        padded_sequences.append(padded_seq)

    targets = np.array(targets).reshape(-1, 1)

    padded_sequences = np.array(padded_sequences)

    padded_sequences = np.nan_to_num(padded_sequences, nan=0.0, posinf=0.0, neginf=0.0)
    targets = np.nan_to_num(targets, nan=0.0, posinf=0.0, neginf=0.0)

    print(f"Target stats - Mean: {targets.mean():.4f}, Std: {targets.std():.4f}")
    print(f"Embedding stats - Mean: {padded_sequences.mean():.4f}, Std: {padded_sequences.std():.4f}")

    return padded_sequences, targets, timestamps, max_sequence_length

def train_evaluate_attention(model, sequences, targets, timestamps, test_split=0.2, epochs=5, batch_size=32):
    """Train and evaluate attention model with proper batching"""

    split_idx = int(len(sequences) * (1 - test_split))

    X_train = sequences[:split_idx]
    y_train = targets[:split_idx]
    X_test = sequences[split_idx:]
    y_test = targets[split_idx:]
    test_timestamps = timestamps[split_idx:]

    print(f"Training samples: {len(X_train)}")
    print(f"Test samples: {len(X_test)}")
    print(f"Sequence shape: {X_train.shape}")
    print(f"Target shape: {y_train.shape}")

    history = model.fit(
        X_train, y_train,
        validation_data=(X_test, y_test),
        epochs=epochs,
        batch_size=batch_size,
        verbose=1
    )

    predictions = model.predict(X_test, batch_size=batch_size)

    return predictions, y_test, test_timestamps, history

def evaluate_metrics(predictions, actuals):
    """Calculate comprehensive evaluation metrics for multi-dimensional targets"""

    mse = mean_squared_error(actuals, predictions)
    mae = mean_absolute_error(actuals, predictions)
    rmse = np.sqrt(mse)

    r2 = r2_score(actuals, predictions)

    cos_similarities = []
    for actual, pred in zip(actuals, predictions):
        cos_sim = np.dot(actual.flatten(), pred.flatten()) / (np.linalg.norm(actual) * np.linalg.norm(pred) + 1e-8)
        cos_similarities.append(cos_sim)
    avg_cos_sim = np.mean(cos_similarities)

    actual_directions = np.sign(actuals)
    pred_directions = np.sign(predictions)
    direction_accuracy = np.mean(actual_directions == pred_directions)

    metrics = {
        'MSE': mse,
        'MAE': mae,
        'RMSE': rmse,
        'R²': r2,
        'Cosine Similarity': avg_cos_sim,
        'Direction Accuracy': direction_accuracy
    }

    return metrics

def run_attention_model(df_news, df_target, window_minutes=30, embedding_dim=768, num_target_variables=1):
    """Complete pipeline for running the attention-based model"""

    sequences, targets, timestamps, max_seq_len = prepare_data_for_attention(
        df_news, df_target, window_minutes
    )

    print(f"Prepared {len(sequences)} sequences with max length {max_seq_len}")

    model = build_model(
        embedding_dim=embedding_dim,
        num_target_variables=num_target_variables,
        attention_hidden_dim=128
    )

    predictions, actuals, test_timestamps, history = train_evaluate_attention(
        model, sequences, targets, timestamps, epochs=10
    )

    metrics = evaluate_metrics(predictions, actuals)

    print("\nModel Performance:")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")

    return model, predictions, actuals, metrics, history
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd

class NewsVolatilityDatasetFixedLength(Dataset):
    def __init__(self, df_news, df_target, window_minutes=30, max_sequence_length=30):
        self.df_news = df_news
        self.df_target = df_target.resample('1min')['BTCUSDT_target'].first().dropna()
        self.window_minutes = window_minutes
        self.max_sequence_length = max_sequence_length

        self.target_timestamps = self.df_target.index.tolist()

    def __len__(self):
        return len(self.target_timestamps)

    def __getitem__(self, idx):
        current_time = self.target_timestamps[idx]

        window_start = current_time - pd.Timedelta(f'{self.window_minutes}min')

        news_in_window = self.df_news[
            (self.df_news.index >= window_start) &
            (self.df_news.index < current_time)
        ]

        if len(news_in_window) == 0:
             news_embeddings = np.zeros((1, 768), dtype=np.float32)
        else:
            try:
                news_embeddings = np.vstack(news_in_window['embedding'].values).astype(np.float32)
            except:
                 news_embeddings = np.zeros((1, 768), dtype=np.float32)

        if len(news_embeddings) > self.max_sequence_length:
            padded_seq = news_embeddings[:self.max_sequence_length]
        else:
            pad_length = self.max_sequence_length - len(news_embeddings)
            padded_seq = np.vstack([news_embeddings, np.zeros((pad_length, news_embeddings.shape[1]), dtype=np.float32)])


        target_volatility = self.df_target.loc[current_time]


        return {
            'news_embeddings': torch.tensor(padded_seq),
            'target_volatility': torch.tensor(target_volatility, dtype=torch.float32)
        }

target_series_for_split = df_target.resample('1min')['BTCUSDT_target'].first().dropna()
split_time_idx = int(0.8 * len(target_series_for_split))
split_timestamp = target_series_for_split.index[split_time_idx]

df_news_train = df_news[df_news.index < split_timestamp]
df_target_train = df_target[df_target.index < split_timestamp]

df_news_val = df_news[df_news.index >= split_timestamp]
df_target_val = df_target[df_target.index >= split_timestamp]


FIXED_SEQ_LEN = 30


train_dataset = NewsVolatilityDatasetFixedLength(df_news_train, df_target_train, window_minutes=30, max_sequence_length=FIXED_SEQ_LEN)
val_dataset = NewsVolatilityDatasetFixedLength(df_news_val, df_target_val, window_minutes=30, max_sequence_length=FIXED_SEQ_LEN)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32)

print(f"Train dataset size (fixed length): {len(train_dataset)}")
print(f"Validation dataset size (fixed length): {len(val_dataset)}")
```

```python
import torch.nn as nn
import torch.optim as optim

class SimpleLSTMModel(nn.Module):
    def __init__(self, embedding_dim, hidden_dim, output_dim, num_layers):
        super(SimpleLSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers

        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)

        self.linear = nn.Linear(hidden_dim, output_dim)

    def forward(self, sequence):
        h0 = torch.zeros(self.num_layers, sequence.size(0), self.hidden_dim).to(sequence.device)
        c0 = torch.zeros(self.num_layers, sequence.size(0), self.hidden_dim).to(sequence.device)

        out, (hn, cn) = self.lstm(sequence, (h0.detach(), c0.detach()))

        output = self.linear(out[:, -1, :])
        return output

embedding_dim = 768
hidden_dim = 128
output_dim = 1
num_layers = 2

model = SimpleLSTMModel(embedding_dim, hidden_dim, output_dim, num_layers)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)


def train_model_pytorch(model, train_loader, val_loader, criterion, optimizer, num_epochs=5):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    print(f"Using device: {device}")

    train_losses = []
    val_losses = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for i, data in enumerate(train_loader):
            inputs = data['news_embeddings'].to(device)
            targets = data['target_volatility'].to(device)

            optimizer.zero_grad()

            outputs = model(inputs)

            targets = targets.view(-1, 1)

            loss = criterion(outputs, targets)

            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            if (i + 1) % 100 == 0:
                print(f'Epoch [{epoch + 1}/{num_epochs}], Step [{i + 1}/{len(train_loader)}], Loss: {loss.item():.4f}')

        epoch_loss = running_loss / len(train_loader)
        train_losses.append(epoch_loss)
        print(f'Epoch [{epoch + 1}/{num_epochs}] finished, Training Loss: {epoch_loss:.4f}')

        model.eval()
        val_running_loss = 0.0
        with torch.no_grad():
            for data in val_loader:
                inputs = data['news_embeddings'].to(device)
                targets = data['target_volatility'].to(device)
                targets = targets.view(-1, 1)

                outputs = model(inputs)
                loss = criterion(outputs, targets)
                val_running_loss += loss.item()

        val_epoch_loss = val_running_loss / len(val_loader)
        val_losses.append(val_epoch_loss)
        print(f'Epoch [{epoch + 1}/{num_epochs}] finished, Validation Loss: {val_epoch_loss:.4f}')

    print('Finished Training')

    return train_losses, val_losses

train_losses, val_losses = train_model_pytorch(model, train_loader, val_loader, criterion, optimizer, num_epochs=5)
```

```python
from sklearn.metrics import r2_score
import numpy as np
import torch

model.eval()

predictions = []
actuals = []

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

with torch.no_grad():
    for data in val_loader:
        inputs = data['news_embeddings'].to(device)
        targets = data['target_volatility'].to(device)

        outputs = model(inputs)

        predictions.extend(outputs.squeeze().cpu().numpy())
        actuals.extend(targets.squeeze().cpu().numpy())

r2 = r2_score(actuals, predictions)

print(f"R² score on the validation set: {r2:.4f}")

from sklearn.metrics import mean_squared_error, mean_absolute_error
mse = mean_squared_error(actuals, predictions)
rmse = np.sqrt(mse)
mae = mean_absolute_error(actuals, predictions)

print(f"MSE on the validation set: {mse:.4f}")
print(f"RMSE on the validation set: {rmse:.4f}")
print(f"MAE on the validation set: {mae:.4f}")
```
